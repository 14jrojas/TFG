```python
import matplotlib.pyplot as plt
import os
```

```python
def extract_data(file_path):
    epochs = []
    losses = []
    accuracies = []
    
    with open(file_path, 'r') as file:
        lines = file.readlines()
        
        for line in lines:
            if 'Epoch' in line:
                epoch = int(line.split('[')[1].split(']')[0])
                epochs.append(epoch)
                
                loss = float(line.split('Loss ')[1].split('\t')[0])
                losses.append(loss)
                
                accuracy = float(line.split('Acc@1 ')[1].split('\n')[0])
                accuracies.append(accuracy)
    
    return epochs, losses, accuracies
```

```python
current_dir = os.getcwd()
file_path1 = os.path.join(current_dir, 'deq', 'output', 'cifar10', 'cls_mdeq_LARGE', 'cls_mdeq_LARGE_train.log')
epochs1, losses1, accuracies1 = extract_data(file_path1)
```
